## scikit-learn으로 회귀분석
### data set 구성하기
1. input_data와 target_data 나누기
2. 각 data를 8:2로 나누어 training set, test set 구성하기
3. training set을 다시 8:2로 나누어 training set, validation set 구성하기


```python
from sklearn.model_selection import train_test_split

# input_dataf(x)와 target_data(y) 나누기
input_data = df.drop(['price'],axis=1)
target_data = df['price']

# 8:2로 나누어 training set, test set 구성
x_train, x_test, y_train, y_test = train_test_split(input_data, target_data, test_size=0.2)

# 8:2로 나누어 training set, validation set 구성
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2)

print('모든 데이터',input_data.shape, target_data.shape)
print('training set',x_train.shape, y_train.shape)
print('valid set',x_valid.shape, y_valid.shape)
print('test set', x_test.shape, y_test.shape)
```

### scikit-learn으로 모델 구현 & 성능 평가
#### LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_train, y_train)
print('train set score', lr.score(x_train, y_train)) #숫자 그대로 이해하지 말고 그냥 1에 가깝다는 식으로 이해
print('valid set score', lr.score(x_valid, y_valid))
print("test set score:", lr.score(x_test, y_test))
```

#### Polynomial Regression

```python
#선형회귀만으로 부족하다고 느낀다면!
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
model = make_pipeline(PolynomialFeatures(2), #2는 2차라는 뜻
                     LinearRegression()).fit(x_train, y_train)
print('train set score',model.score(x_train, y_train))
print('valid set score',model.score(x_valid, y_valid))
print("test set score:", model.score(x_test, y_test))
```

```python
model = make_pipeline(PolynomialFeatures(3),
                     LinearRegression()).fit(x_train, y_train)
print('train set score',model.score(x_train, y_train))
print('valid set score',model.score(x_valid, y_valid))
print("test set score:", model.score(x_test, y_test))
```

#### DecisionTree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
tree= DecisionTreeRegressor().fit(x_train, y_train)
print('train set score:', tree.score(x_train, y_train))
print('valid set score:', tree.score(x_valid, y_valid))
print("test set score:", tree.score(x_test, y_test))
```

```python
# 그래프로 표현하기
train_score=[]
valid_score=[]
depth_range=range(1,25)
for depth in depth_range:
    tree=DecisionTreeRegressor(max_depth=depth).fit(x_train, y_train)
    train_score.append(tree.score(x_train, y_train))
    valid_score.append(tree.score(x_valid, y_valid))
plt.plot(depth_range, train_score)
plt.plot(depth_range, valid_score)
plt.legend(['train','Valid'])
plt.xlabel('Max Depth')
plt.ylabel('Score')
plt.show()
```
