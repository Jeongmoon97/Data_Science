# matplotlib

파이썬으로 기본적인 차트들을 쉽게 그릴 수 있도록 도와주는 데이터 시각화 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
# 그래프의 결과를 출력 세션에 나타나도록 함
%matplotlib inline
```

### 기본 그래프

```python
plt.figure
           #x축       y축
plt.plot([1,2,3], [90,100,50])
plt.show()
```

```python
plt.figure(figsize=(10,6))

# 데이터, 축이름, 범례 추가
plt.plot(['Math','Korean','English'], [90,100,50])
plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm Exam')

# 데이터 추가
plt.plot(['Math','Korean','English'], [70,50,100])

# 범례
plt.legend(['Choi','Kim'])

plt.show()
```

```python
plt.figure(figsize=(10,6))

# 범례를 추가하는 다른 방법
plt.plot(['Math','Korean','English'], [90,100,50], label='Choi')

plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm Exam')

plt.plot(['Math','Korean','English'], [70,50,100], label='Kim')

# 범례 추가
plt.legend()

plt.show()
```

### 차트 스타일 변경


```python
plt.figure(figsize=(10,6))

# 색상, 마커, 라인스타일
plt.plot([1,2,3],[100,90,50], color = 'red', markerfacecolor = 'blue',
         marker='.', markersize=30, linestyle='--')

# 격자무늬
plt.grid(True) 

# x축 최소/최대, y축 최소/최대
plt.axis([0,5,0,150]) 
plt.show()
```

```python
plt.figure(figsize=(10,6))

# 선 굵기
plt.plot(['Math','Korean','English'], [90,100,50], lw=3)
plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm Exam')

# 색상
plt.plot(['Math','Korean','English'], [70,50,100], 'r')

plt.legend(['Choi','Kim'])

plt.show()
```

```python
plt.plot([1,2,3],[30,40,30], color='black')
plt.plot([7,8,9],[30,40,30], color='black')
plt.plot([2,8],[30,30],linestyle='',marker='.', markersize=70, color='black')
plt.plot([4,5,6],[20,10,20], color='red')
```

### 그 외 다양한 그래프

```python
x = ['Math','English','Korean']
y = [90, 100, 80]
plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Miderm Exam')
plt.legend('Choi')
plt.bar(x,y, color='pink', width = 0.5)
plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.scatter(t,y, marker='>')
plt.show()
```

```python
# x축 값인 r에 따라 색상을 바꾸는 colormap, 마커의 크기를 지정하는 s
colormap = t

plt.figure(figsize=(10,6))
plt.scatter(t,y, s = 50, c = colormap, marker='>')
plt.colorbar()
plt.show()
```

```python
import numpy as np
# loc는 평균값, scale은 표준편차 지정
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
```

```python
plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```

# Seaborn

matplotlib를 기반으로 만들어져 통계 데이터 시각화에 최적화된 라이브러리

```python
import pandas as pd
train = pd.read_csv('data/titanic_train.csv')
train.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# '나이'와 '생존' 요소를 기준으로 산포도 그리기
sns.relplot(data=train, x='Age', y='Survived')

# 범주 데이터 넣기, 사이즈 조절
sns.relplot(data=train, x='Age', y='Survived', hue='Sex', aspect=3) # aspect->가로길이와 세로길이 비율
```

```python
#Categorical Scatter Plot 그려보기
sns.catplot(data=train, x='Pclass',y='Fare', aspect=4)
sns.catplot(data=train, x='Pclass',y='Fare', jitter=False, aspect=4)
sns.catplot(data=train, x='Pclass',y='Fare', aspect=4, hue='Sex')
```

```python
#박스플롯 그려보기
sns.catplot(data=train, x='Pclass',y='Fare', aspect=1, kind='box')
sns.catplot(data=train, x='Pclass',y='Fare', aspect=1, kind='box', hue='Sex')
```

```python
#pair plot 기능 사용하기
sns.pairplot(data=train, hue='Sex',
            x_vars=['Pclass','Age','Fare'],
            y_vars=['Survived','Age','Fare'], height=3)
```
